–ø–æ—ç—Ç–æ–º—É –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ —Å–µ—á–µ–Ω–∏–µ $Œæ(t) ‚àà Be(1 ‚àí t), t ‚àà [0, 1].$




–ó–∞–ø–∏—à–µ–º –ø–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é —Ñ–æ—Ä–º—É–ª—É –¥–≤—É–º–µ—Ä–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

–í—ã—á–∏—Å–ª–∏–º —ç—Ç—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–≤ –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è $x_1$

 –ù–∞—á–Ω–µ–º —Å –∫—Ä–∞–π–Ω–∏—Ö —Å–ª—É—á–∞–µ–≤, –∫–æ–≥–¥–∞ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –∏–∑ $x_1 $ –∏–ª–∏ $ x_2$ –ø—Ä–µ–≤—ã—Å–∏—Ç –µ–¥–∏–Ω–∏—Ü—É –∏–ª–∏
–º–µ–Ω—å—à–µ –Ω—É–ª—è. –ê–Ω–∞–ª–∏–∑–∏—Ä—É—è –ø–æ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–ª—É—á–∞–∏, –º–æ–∂–Ω–æ –ø—Ä–∏–π—Ç–∏ –∫ –≤—ã—Ä–∞–∂–µ–Ω–∏—é

–ü—Ä–∏ –ø–æ–¥—Å—á–µ—Ç–µ –ø–µ—Ä–≤—ã—Ö —á–µ—Ç—ã—Ä–µ—Ö

–ü—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è
—ç—Ç–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ, —Ç—Ä–µ–±—É–µ—Ç—Å—è –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–º –≤–∏–¥–æ–º —Ñ—É–Ω–∫—Ü–∏–∏ Œæ(œâ, t):

$\min(t_1, t_2) = t_1t_2,$
—á—Ç–æ –≤–µ—Ä–Ω–æ –ª–∏—à—å –≤ —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –∏–∑ –∑–Ω–∞—á–µ–Ω–∏–π $t_1$ –∏–ª–∏ $t_2$
—Ä–∞–≤–Ω–æ –Ω—É–ª—é –∏–ª–∏ –µ–¥–∏–Ω–∏—Ü–µ. 

–í —Ç–∞–∫–∏—Ö —Å–ª—É—á–∞—è—Ö $Œæ(t_1)$ –∏ $Œæ(t_2)$ —è–≤–ª—è—é—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ —Å–ª—É—á–∞–π–Ω—ã–º–∏ –≤–µ–ª–∏—á–∏–Ω–∞–º–∏.

$t_1 \neq 0, 1$ –∏ $t_2 \neq 0, 1,$ —Å–µ—á–µ–Ω–∏—è $Œæ(t_1) $–∏ $Œæ(t_2)$ —è–≤–ª—è—é—Ç—Å—è –∑–∞–≤–∏—Å–∏–º—ã–º–∏ —Å–ª—É—á–∞–π–Ω—ã–º–∏ –≤–µ–ª–∏—á–∏–Ω–∞–º–∏. 

**–û–ø—Ä.** –ü—É—Å—Ç—å {Œæ(t), t ‚àà T} –∏ {Œ∑(t), t ‚àà T} ‚Äì –¥–≤–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –ø—Ä–æ—Ü–µ—Å—Å–∞, –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ –Ω–∞ –æ–¥–Ω–æ–º –∏ —Ç–æ–º –∂–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ (‚Ñ¶, F, P) –∏ –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –æ–¥–Ω–æ–º –∏ —Ç–æ–º –∂–µ –∏–∑–º–µ—Ä–∏–º–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ.

–ü—É—Å—Ç—å $ F(x)$ —è–≤–ª—è–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –ü–æ–∫–∞–∂–µ–º –∫–∞–∫, –∏–º–µ—è –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –≤—ã–±–æ—Ä–∫–∏ –∏–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–≥–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –ø–æ–ª—É—á–∏—Ç—å –≤—ã–±–æ—Ä–∫—É –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –∑–∞–¥–∞–≤–∞–µ–º–æ–≥–æ —Ñ—É–Ω–∫—Ü–∏–µ–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $ F(x)$.

–°—Ç—Ä–æ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
–ï—Å–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è $  {\displaystyle F:\mathbb {R} \to [0,1]} $ —Å—Ç—Ä–æ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –Ω–∞ –≤—Å–µ–π –æ–±–ª–∞—Å—Ç–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Ç–æ –æ–Ω–∞ –±–∏–µ–∫—Ç–∏–≤–Ω–∞, –∞ —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∏–º–µ–µ—Ç –æ–±—Ä–∞—Ç–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é ${\displaystyle F^{-1}:[0,1]\to \mathbb {R} }.$

–ü—É—Å—Ç—å ${\displaystyle U_{1},\ldots ,U_{n}\sim U[0,1]} $ ‚Äî –≤—ã–±–æ—Ä–∫–∞ –∏–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–≥–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
–¢–æ–≥–¥–∞ $ X_{1},\ldots ,X_{n}, $ –≥–¥–µ$ {\displaystyle X_{i}=F^{-1}(U_{i}),\;i=1,\ldots ,n},$ ‚Äî –≤—ã–±–æ—Ä–∫–∞ –∏–∑ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–µ–≥–æ –Ω–∞—Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.



–ü—É—Å—Ç—å —Ç—Ä–µ–±—É–µ—Ç—Å—è —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –≤—ã–±–æ—Ä–∫—É –∏–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º $ \lambda >0.$ 

$$f_X(x) = \begin{cases} \lambda e^{-\lambda x}, x\ge 0  \\
0, x<0
\end{cases}$$

–§—É–Ω–∫—Ü–∏—è —ç—Ç–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $ {\displaystyle F(x)=1-e^{-\lambda x}}$ —Å—Ç—Ä–æ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç, –∏ –µ—ë –æ–±—Ä–∞—Ç–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –∏–º–µ–µ—Ç –≤–∏–¥ $ {\displaystyle F^{-1}(x)=-{\frac {1}{\lambda }}\,\ln(1-x)}. $ 

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –µ—Å–ª–∏ ${\displaystyle U_{1},\ldots ,U_{n}}$ ‚Äî –≤—ã–±–æ—Ä–∫–∞ –∏–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–≥–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Ç–æ $ X_{1},\ldots ,X_{n},$ –≥–¥–µ

${\displaystyle X_{i}=-{\frac {1}{\lambda }}\ln(1-U_{i}),\;i=1,\ldots ,n}$
‚Äî –∏—Å–∫–æ–º–∞—è –≤—ã–±–æ—Ä–∫–∞ –∏–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

```python
import math

def my_exp(lmbda=1.0):
    return (-1 / lmbda)*math.log(1 - random.random())

plt.hist([my_exp() for i in range(10000)], 50, normed=True)
plt.title('Samples from hand rolled exponential function')
None
```

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ Box-Muller 

**–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ Box-Muller** - –º–µ—Ç–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—É—Ç–µ–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≤—ã–±–æ—Ä–∫–∏ –∏–∑ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, $u_1$ –∏ $u_2$ - –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã –Ω–∞ $[0,1]$.

–ü–æ–ª–æ–∂–∏–º,
$$ z_1 = \sqrt{(-2 \log(u_1))}\cos(2\pi u_2) $$
 $$ z_2 = \sqrt{(-2 \log(u_1))}\sin(2\pi u_2)$$

```python
from numpy import random, sqrt, log, sin, cos, pi
from pylab import show,hist,subplot,figure

# transformation function
def gaussian(u1,u2):
  z1 = sqrt(-2*log(u1))*cos(2*pi*u2)
  z2 = sqrt(-2*log(u1))*sin(2*pi*u2)
  return z1,z2

u1 = random.rand(1000)
u2 = random.rand(1000)
z1,z2 = gaussian(u1,u2)
```

```python

# plotting the values before and after the transformation
figure()
subplot(221) # the first row of graphs
hist(u1)     # contains the histograms of u1 and u2 
subplot(222)
hist(u2)
subplot(223) # the second contains
hist(z1)     # the histograms of z1 and z2
subplot(224)
hist(z2)
show()
```

$$\xi =\mu +\sigma z$$

```python
pic.shape
```

```python
plt.imshow(pic, cmap='gray')
```

```python
def s_filter(image, fil):
    pr_im = np.zeros((image.shape[0]-2, image.shape[1]-2))
    for i in range(image.shape[0]-2):
        for j in range(image.shape[1]-2):
            pr_im[i, j] = np.sum(image[i:i+3, j:j+3] * fil)
    return pr_im
```

```python
new_pic = s_filter(pic, sobel_filter)
```

```python
plt.imshow(new_pic, cmap='gray')
```

```python
plt.imsave('new_pic.png', new_pic, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input_image = plt.imread('skyscrapers.png')[:,:,0]
```

```python
plt.imshow(input_image, cmap='gray')
```

```python
def max_pooling(inp, size):
    pr_pic = np.zeros((inp.shape[0]//size, inp.shape[1]//size)).astype('float32')
    for i in range(inp.shape[0]//size):
        for j in range(inp.shape[1]//size):
            pr_pic[i, j] = np.max(
                inp[size*i:size*i+size, 
                    size*j:size*j+size])
    return pr_pic
```

```python
max_pooling(input_image, 3)
```

```python
plt.imshow(max_pooling(input_image, 3), cmap='gray')
```

```python
plt.imsave('pooled_skyscrapers.png', max_pooling(input_image, 3), cmap='gray')
```

```python
plt.imsave('pooled_skyscrapers_x10.png', max_pooling(input_image, 10), cmap='gray')
```

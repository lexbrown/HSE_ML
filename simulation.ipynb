```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

np.linspace(0, 10, 21)
```

```python
points = 81
np.random.seed(42)
aa = np.random.randn(points)

plt.plot(np.linspace(0, 10, points), aa)
plt.plot(np.linspace(0, 10, points), np.zeros(points), color='red')
```

```python
sns.distplot(aa)
```

```python
plt.plot(np.linspace(0, 10, points), np.cumsum(aa))
plt.plot(np.linspace(0, 10, points), np.zeros(points), color='red')
```

```python
def reverse():
    n = int(input())
    if n != 0:
        reverse()
        print(n)
```

```python
del reverse
```

```python
def lol():
    n = int(input())
    if n != 0:
        if n %2 == 0:
            print(n)
        lol()
        if n %2 != 0:
            print(n)
```
